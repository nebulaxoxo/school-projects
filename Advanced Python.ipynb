# Matplotlib


Matplotlib is a Python 2D plotting library that we can use to produce high quality data visualization. It is highly usable (as you will soon find out), you can create simple and complex graphs with just a few lines of codes!

```python
#importing the package as plt
import matplotlib.pyplot as plt
import numpy as np
```

We import numpy as it makes the working with numbers easy which is being used in plotting graphs.

Let us start with a simple graph by using plot() function. plot() is a versatile command, and will take an arbitrary number of arguments.

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0, 1, 2, 3].

We can specify the x values separately. Let us try by mentioning the values. 

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

# Giving label to the x and y axis

We can label the xaxis and the yaxis which makes it easier to understand the data displayed in the graph.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel("X_numbers")
plt.ylabel('Y_numbers')
plt.show()
```

we can also add title to the graph.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("My first Graph")
plt.xlabel("X_numbers")
plt.ylabel('Y_numbers')
plt.show()
```

# Change size of the graph

```python
plt.figure(figsize=(10,10))

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("My first Graph")
plt.xlabel("X_numbers")
plt.ylabel('Y_numbers')
plt.show()
```

# Creating a Subplot

pyplot.subplots() creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created.


Let us try by plotting 4 different graphs in a single plot with the help of subplots. For this let us first assume the data for 4 different graphs

```python
#Here we are taking same value of x for all the graphs
y1 = [0,2,4,6]
y2 = [11,13,15,17]
y3 = [12,22,32,52]
y4 = [0,5,10,15]
x = [1,2,3,4]

#Let's first create subplot

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x,y3)
axs[1, 1].plot(x, y4)

plt.show()
```

This divides the plot into a matrix and each plot in the matrix can be accesses using the indexing of that matrix as used in the above code.
